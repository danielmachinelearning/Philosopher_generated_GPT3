```python
import os
import openai
import random

openai.api_key = "<API_KEY>"

def openFiles(textChunkSize):
  lengths = []
  chunks = []

  with open('wittgenstein.txt', 'r') as file:
    wittgenstein_text = file.read()

  with open('heidegger.txt', 'r') as file:
      heidegger_text = file.read()

  with open('hegel.txt', 'r') as file:
      hegel_text = file.read()

  len_wittgenstein = len(wittgenstein_text)
  len_heidegger = len(heidegger_text)
  len_hegel = len(hegel_text)

  lengths = [len_wittgenstein, len_heidegger, len_hegel]

  chunks_wittgenstein = int(len_wittgenstein / textChunkSize)
  chunks_heidegger = int(len_heidegger / textChunkSize)
  chunks_hegel = int(len_hegel / textChunkSize)

  chunks = [chunks_wittgenstein, chunks_heidegger, chunks_hegel]
  texts = [wittgenstein_text, heidegger_text, hegel_text]

  return lengths, chunks, texts

def getTextPiece(textList, chunksValue, textChunkSize, i):
  if i == chunksValue:
    given_text1 = textList[i*textChunkSize:]
  elif i < chunksValue:
    given_text1 = textList[i*textChunkSize:(i+1)*textChunkSize]
  else:
    list1 = range(chunksValue)
    rand_num = random.choice(list1)
    given_text1 = textList[rand_num*textChunkSize:(rand_num+1)*textChunkSize]

  return given_text1

def setupInputText(selected_text, final_text, past_length, i):
  given_text = ''
  
  for k in selected_text:
    given_text += k

  if i == 0:
    input_text = given_text
  else:
    input_text = final_text[-past_length:-1] + given_text

  return input_text, given_text




```

```python
final_text = ''
response_text = []

chunk_text_size = 512
max_token_size = 1024
past_lookup = max_token_size + 3*chunk_text_size

lengths, chunks, texts = openFiles(512)

for i in range(30):
  given_text1 = getTextPiece(texts[0], chunks[0], chunk_text_size, i)
  given_text2 = getTextPiece(texts[0], chunks[0], chunk_text_size, i)
  given_text3 = getTextPiece(texts[0], chunks[0], chunk_text_size, i)

  selected_text = [given_text1, given_text2, given_text3]
  random.shuffle(selected_text)

  input_text, given_text = setupInputText(selected_text, final_text, past_lookup, i)

  response = openai.Completion.create(
    model="text-davinci-002",
    prompt=input_text,
    temperature=0.6,
    max_tokens=max_token_size,
    top_p=1,
    frequency_penalty=1,
    presence_penalty=1,
  )
  
  final_text = final_text + given_text + response['choices'][0]['text']
  overall_response = response['choices'][0]['text']

  response_text.append(overall_response)
```

```python
for i,j in zip(response_text,range(len(response_text))):
  print("Statement %s: " % j)
  print(i)
  print("\n")
```
